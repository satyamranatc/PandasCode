```python
x = ["A", "B", "C", "D", "E"]
y = [10, 20, 30, 40, 50]

plt.bar(x,y, color = "teal", width=0.1)
# plt.barh(x,y, color = "teal", height=0.1) #Horizontal

plt.xlabel("Name")
plt.ylabel("Age")
plt.title("Age vs Name")

plt.show()
```

```python
x = ["A", "B", "C", "D", "E"]
y = [10, 20, 30, 40, 50]

plt.bar(x,y, color = "teal", width=0.1)
plt.plot(x,y, color = "red") 

plt.grid()
plt.xlabel("Name")
plt.ylabel("Age")
plt.title("Age vs Name")

plt.show()
```

```python
x = ["A","B","C","D"]
y = [10,20,30,25]


plt.subplot(1,2,2)
plt.plot(x,y,color = "#510DF7",marker = "o",linestyle = "dashed")
plt.xlabel("Name")
plt.ylabel("Age")
plt.title("Age vs Name")

x2 = ["A2","B2","C2","D2"]
y2 = [12,22,33,29]


plt.subplot(1,2,1)
plt.plot(x2,y2,color = "#510DF7",marker = "o",linestyle = "dashed")
plt.xlabel("Name")
plt.ylabel("Age")
plt.title("Age vs Name")

plt.show()
```
